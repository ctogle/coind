```python
import matplotlib.pyplot as plt
import numpy as np
from coind.training import train


ticker_logs = (
    './data/stream.7.log',
    './data/stream.6.log',
    './data/stream.longer.5.log',
    './data/stream.longer.3.log',
    './data/stream.longer.2.log',    
)

epochs = 5
models, train_losses, val_losses, accuracies = train(epochs, './.models/', ticker_logs)

f, ax = plt.subplots(1, 1, figsize=(10, 3))
for y in (train_losses, val_losses, accuracies):
    y = np.array(y) / max(y)
    x = np.linspace(0, epochs + 1, len(y))
    ax.plot(x, y)
f.show()
```
