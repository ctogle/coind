```python
import torch
from coind.inference import inference


class agent:
    
    def __init__(self, model, source=None):
        self.model = torch.load(model) if isinstance(model, str) else model
        print(self.model.n_features)
        self.source = source

    def live(self):
        print('agent operating in product market:', self.model.products)
        for bull, prices in inference(self.model, 3, 600, self.source):
            print(bull)
            print(prices)
```

```python
model = './.models/oracle.pt'
source = './data/stream.6.log'

a = agent(model, source)
a.live()
```
